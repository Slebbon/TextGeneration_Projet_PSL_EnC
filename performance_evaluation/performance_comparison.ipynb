Put both in pipeline

```python
text_generator = pipeline("text-generation", model=gp2_finetuned_model, tokenizer=gp2_finetuned_tokenizer, device=0, pad_token_id=50256)
base_text_generator = pipeline("text-generation", model=base_gpt2_model, tokenizer=base_gpt2_tokenizer,device=0, pad_token_id=50256)
```

```python
prova = text_generator(['Thou be','Hello!'], max_length=256,truncation=True)
```

```python
def get_answer(prompt, print_answer=False, save=True):
  if isinstance(prompt, str):
    prompt = [prompt]
  base_answer= base_text_generator(prompt, max_new_tokens=256, truncation=True)
  answer = text_generator(prompt, max_new_tokens=256,truncation=True)
  base_answer = list(map(lambda x: x[0]['generated_text'], base_answer))
  answer = list(map(lambda x: x[0]['generated_text'], answer))
  if print_answer:
     print(f'base model: {base_answer} \n')
     print(f'finetuned model: {answer} \n')

  if save:
    return base_answer, answer
  else:
    return None
```

# Mistral

```python
from huggingface_hub import notebook_login
notebook_login()
```

```python
base_model_id = "mistralai/Mistral-7B-v0.1"
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

base_mistral_model = AutoModelForCausalLM.from_pretrained(base_model_id, quantization_config=bnb_config)
base_mistral_tokenizer = AutoTokenizer.from_pretrained(
    base_model_id,
    padding_side="left",
    add_eos_token=True)


f_mistral_model = AutoModelForCausalLM.from_pretrained(base_model_id, quantization_config=bnb_config)
```

```python
from peft import prepare_model_for_kbit_training
f_mistral_model = prepare_model_for_kbit_training(f_mistral_model)
```

```python
from peft import PeftModel, PeftConfig
```

```python
f_mistral_model = PeftModel.from_pretrained(f_mistral_model, '/content/drive/MyDrive/mistral_lora_shrump', is_trainable=True)
f_mistral_tokenizer = AutoTokenizer.from_pretrained('/content/drive/MyDrive/mistral_lora_shrump')
```

```python
base_mistral_generator = pipeline("text-generation", model=base_mistral_model, tokenizer=base_mistral_tokenizer,pad_token_id=2)
mistral_generator = pipeline("text-generation", model=f_mistral_model, tokenizer=f_mistral_tokenizer,pad_token_id=2)
```

```python
def get_answer_mistral(prompt, print_answer=False, save=True, max_new_tokens=128):
  if isinstance(prompt, str):
    prompt = [prompt]
  base_answer= base_mistral_generator(prompt, max_new_tokens=max_new_tokens, truncation=True)
  answer = mistral_generator(prompt, max_new_tokens=max_new_tokens,truncation=True)

  base_answer = list(map(lambda x: x[0]['generated_text'], base_answer))
  answer = list(map(lambda x: x[0]['generated_text'], answer))
  if print_answer:
    print(f'base model: {base_answer} \n')
    print(f'finetuned model: {answer} \n')

  if save:
    return base_answer, answer
```

```python
get_answer_mistral('O Lord', print_answer=True, save=False)
```
